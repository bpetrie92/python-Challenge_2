```python
import csv
from pathlib import Path

menu_filepath = Path('../../Instructions/PyRamen/Resources/menu_data.csv')
sales_filepath = Path('../../Instructions/PyRamen/Resources/sales_data.csv')

menu = []
sales = []

with open(menu_filepath, "r") as csvfile:
    menu_file = csv.reader(csvfile, delimiter = ",")
    header1 = next(menu_file)
#    print(header1)
    
    for row in menu_file:
        menu.append(row)
#        print(row)    
#print(menu)


with open(sales_filepath, "r") as csvfile:
    sales_file = csv.reader(csvfile, delimiter = ",")
    header2 = next(sales_file)
#    print(header2)

    for row in sales_file:
        sales.append(row)
#        print(row[0:5])
    
#print(len(sales))

report = {}
row_count = 0

    # @TODO: Initialize sales data variables
sales_data=['Quantity', 'Menu_item']
#print(sales_data)

    # @TODO:
    # If the item value not in the report, add it as a new entry with initialized metrics
    # Naming convention allows the keys to be ordered in logical fashion, count, revenue, cost, profit

for row in sales:
    quantity = int(row[3])
    menu_item = (row[4])
    
    if menu_item not in report.keys():
        report[menu_item] = {'count':0, 'revenue':0, 'cost':0, 'profit':0}
        
    for x in menu:
        item = x[0]
        price = float(x[3]) 
        cost = float(x[4])
        profit = price -cost
        
        if menu_item == item:
            print('Item found:')
            print(item)
            print(price)
            print(cost)
            print(profit)
            
            report[menu_item]['count'] += quantity
            report[menu_item]['revenue'] += price * quantity
            report[menu_item]['cost']  += cost * quantity
            report[menu_item]['profit'] += profit * quantity
            
        else:
            print(f'no match for {item}')
            
    row_count +=1          
print(f" total transactions: {row_count}")        
        
 




# @TODO: Write out report to a text file (won't appear on the command line output)
```

```python
with open('report.txt', 'w') as file:
    for y, z in report.items():
        file.write(f'{y} {z}\n')
```
